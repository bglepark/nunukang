```python
# data11.to_csv('test.csv' , index=False)

x_min = []
y_min = []
b_w = []
b_h = []


a = data11['bbox'].values

for i in range(len(data11)):
    x_min.append(a[i][0])
    y_min.append(a[i][1])
    b_w.append(a[i][2])
    b_h.append(a[i][3])

# data11.head()
```

```python
y_min
```

```python
# xmin: top-left x coordinate,
# ymin: top-left y coordinate,
# w: bounding box width,
# h: bounding box height,
# w_img: image width,
# h_img: image height

# xcenter = (xmin + w/2) / w_img
# ycenter = (ymin + h/2) / h_img
# w = w / w_img
# h = h / h_img
```

```python
x_center = []
y_center = []
w = []
h = []

for i in range(len(x_min)):
    x_center.append((x_min[i]+(b_w[i]/2))/2704)
    y_center.append((y_min[i]+(b_h[i]/2))/1520)
    w.append(b_w[i]/2704)
    h.append(b_h[i]/1520)
    
```

```python
print(len(x_center))
print(len(y_center))
print(len(w))
print(len(h))

y_center
```
