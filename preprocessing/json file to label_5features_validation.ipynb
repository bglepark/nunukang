```python
data1 = pd.json_normalize(file['images'])
data1.head()
```

```python
data2 = pd.json_normalize(file['annotations'])
data2.head()
```

```python
data2['fish_size'] = None

print(data2['bbox'][0][2]*0.026458333)
print(data2['bbox'][1][2]*0.026458333)

print(len(data2['bbox']))
```

```python
for i in range(len(data2['bbox'])):
    data2['fish_size'][i] = data2['bbox'][i][2]*0.026458333
    
data2['fish_size'].head()
```

```python
data3 = pd.concat([data1 , data2] , axis=1)
data3.head()
```

```python
data3 = data3.drop(columns=['light', 'date_captured' ,'diseases_bbox' , 'attribute', 'iscrowd'])
```

```python
data3 = data3.drop(columns=['image_id' , 'keypoints' , 'gd' , 'id' , 'id'])
```

```python
# 질병이 없는 것만 다시 추림

data4 = data3[data3['diseases_exist'] == False]
data4
```

```python
data4 = data4.drop(columns=['diseases_desc' , 'diseases_exist'])
```

```python
# id : 1(넙치(광어)) Olive flounder 금지체장 : 35
# id : 2(우럭(조피볼락)) Korea rockfish 23
# id : 3(참돔) Red seabream 금지체장 : 24
# id : 4(감성돔) Black porgy 25
# id : 5(돌돔) Rock bream 24
data4
```
